Importing Libraries


```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

Reading Csv's


```
wodi=pd.read_csv("/content/CricinfoICC Women's ODI Team Rankings.csv")
odi = pd.read_csv("/content/CricinfoICC ODI Championship.csv")
odi
```

```
wodi
```

Fuction which trims team names column to 3 letters


```
def trim(x):
  a=x[:3]
  return a
```

```
odi['team']=odi['team'].apply(trim)
wodi['team']=wodi['team'].apply(trim)
odi
```

```
wodi
```

Increasing figure Size and Ploting bar Graph of teams vs Points

```
f,ax=plt.subplots(figsize=(20,10))
sns.barplot('team','points',data=odi)
```

Plotting the bargraph for team vs rating

```
f,ax=plt.subplots(figsize=(20,10))
ax=sns.barplot('team','rating',data=odi)
```

Plotting horizontal barplot for rationg vs teams

```
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('rating','team',data=odi)
```

Plotting the horizontal barplot for matches vs team


```
f,ax=plt.subplots(figsize=(20,10))
ax=sns.barplot('matches','team',data=odi)
```

Concatinating two data sets with gender as new Column

```
odi['gender']="male"
wodi['gender']='female'
Todi=pd.concat([wodi,odi],ignore_index=True)
Todi
```

Plotting team vs points with hue as Gender

```
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('team','points',hue='gender',data=Todi)
```
